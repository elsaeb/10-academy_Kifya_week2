```python
# Aggregate information per user
per_user_aggregation = data.groupby('IMSI').agg({
    'Bearer Id': 'count',
    'Dur. (ms)': 'sum',
    'Total DL (Bytes)': 'sum',
    'Total UL (Bytes)': 'sum'
}).reset_index()

application_columns = ['Social Media DL (Bytes)', 'Social Media UL (Bytes)', 'Email DL (Bytes)','Email UL (Bytes)', 'Youtube DL (Bytes)','Youtube UL (Bytes)', 'Netflix DL (Bytes)'
,'Netflix UL (Bytes)', 'Gaming DL (Bytes)','Gaming UL (Bytes)', 'Other DL (Bytes)', 'Other UL (Bytes)','Google DL (Bytes)','Google UL (Bytes)']
per_user_aggregation[application_columns] = data.groupby('IMSI')[application_columns].sum().reset_index(drop=True)

per_user_aggregation.head()

```

```python
# Outlier handling 
Q1 = per_user_aggregation['Dur. (ms)'].quantile(0.25)
Q3 = per_user_aggregation['Dur. (ms)'].quantile(0.75)
IQR = Q3 - Q1

# Remove outliers
per_user_aggregation = per_user_aggregation[(per_user_aggregation['Dur. (ms)'] >= (Q1 - 1.5 * IQR)) & (per_user_aggregation['Dur. (ms)'] <= (Q3 + 1.5 * IQR))]
```

```python
per_user_aggregation['total_data_volume']=per_user_aggregation['Total DL (Bytes)'] + per_user_aggregation['Total UL (Bytes)']
```

```python
# Segment users into deciles
per_user_aggregation['decile_class'] = pd.qcut(per_user_aggregation['Dur. (ms)'], 10, labels=False)

# Calculate total data volume per decile class
decile_data = per_user_aggregation.groupby('decile_class')['Dur. (ms)'].sum().reset_index()
```

```python
# Aggregate metrics per customer ID (MSISDN)
user_engagement = data.groupby('MSISDN/Number').agg({
    'Dur. (ms)': 'sum',
    'Bearer Id': 'count',
    'Total DL (Bytes)': 'sum',
    'Total UL (Bytes)': 'sum' }).reset_index()

top_10_duration = user_engagement['Dur. (ms)'].nlargest(10)
top_10_sessions = user_engagement['Bearer Id'].nlargest(10)
top_10_traffic_dl = user_engagement['Total DL (Bytes)'].nlargest(10)
top_10_traffic_ul = user_engagement['Total UL (Bytes)'].nlargest(10)


print("Top 10 customers by session duration:\n", top_10_duration)
print("Top 10 customers by session frequency:\n", top_10_sessions)
print("Top 10 customers by total download size:\n", top_10_traffic_dl)
print("Top 10 customers by total upload size:\n", top_10_traffic_ul)
```

```python
user_engagement['total_data_volume']=user_engagement['Total DL (Bytes)'] + user_engagement['Total UL (Bytes)']
```

```python
from sklearn.preprocessing import MinMaxScaler

# Normalize the metrics
scaler = MinMaxScaler()
user_engagement[['Dur. (ms)', 'Bearer Id', 'total_data_volume']] = scaler.fit_transform(user_engagement[['Dur. (ms)', 'Bearer Id', 'total_data_volume']])
```

```python
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings('ignore')

# Run K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=0)
user_engagement['cluster'] = kmeans.fit_predict(user_engagement[['Dur. (ms)', 'Bearer Id', 'total_data_volume']])

# Analyze clusters
cluster_summary = user_engagement.groupby('cluster').agg({
    'Dur. (ms)': ['min', 'max', 'mean', 'sum'],
    'Bearer Id': ['min', 'max', 'mean', 'sum'],
    'total_data_volume': ['min', 'max', 'mean', 'sum']
}).reset_index()

print(cluster_summary)
```
